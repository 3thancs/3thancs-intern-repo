```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Load the dataset and read the dataset
CarData_path = r"C:\Users\ASUS\Desktop\Data Science Swinburne\Professional Internship\3thancs-intern-repo\Cars Datasets 2025.csv"
CarData = pd.read_csv(CarData_path, encoding='latin-1')

# View the descriptive statistics of the dataset for 
df = ['HorsePower', 'Total Speed', 'Performance(0 - 100 )KM/H', 'Cars Prices', 'Seats', 'Torque']
print("Numeric statistics:")
print(CarData[df].describe())
```

```python
# Check for missing values
print("\n Missing values in each column:")
print(CarData.isnull().sum())
```

```python
#Drop rows of missing values of data 
CarData = CarData.dropna()
print(CarData.isnull().sum())
```

```python
# Using regplot for a simple X and Y plot with regression line
sns.regplot(x='HorsePower', y='Cars Prices', data=CarData)
plt.title('HorsePower vs Cars Prices')
plt.xlabel('HorsePower')
plt.ylabel('Cars Prices')
plt.show()
```
